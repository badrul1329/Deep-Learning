```python
# Custom Split
xtrain = []
ytrain = []
length = len(df)
for i in range(0,len(df)):
    xtrain.append(list(df.values[i,:4]))
    ytrain.append(df.values[i,4])
    
len(xtrain),len(ytrain)
```

```python
# Normalization
std = StandardScaler()
xtrain_scaler = std.fit_transform(xtrain)
xtrain_scaler[:5]
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (12,5))
ax1.set_title("Before Scalling of Train Data")
ax2.set_title("After Scalling of Train Data")
lebels = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i in range(0,4):
    before = []
    after = []
    # before
    for beforeRow in xtrain:
        before.append(beforeRow[i])
    # after
    for afterRow in xtrain_scaler:
        after.append(afterRow[i])
    # Kernel Distribution Estimation Plot for Pre-existing axes ax1/ax2
    sns.kdeplot(before, ax = ax1, label = lebels[i])
    sns.kdeplot(after, ax = ax2, label = lebels[i])
plt.suptitle("Comparison")
plt.show()
```

```python
# Comparison using subplot
plt.figure(figsize=(16,6))

# plt.subplot(row, col, no)
plt.subplot(1, 2, 1)
plt.title("Before Scaling of Train Data")
plt.plot(xtrain)

plt.subplot(1, 2, 2)
plt.title("After Scaling of Train Data")
plt.plot(xtrain_scaler)

plt.legend(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
plt.suptitle("Comparison")
plt.show()
```
